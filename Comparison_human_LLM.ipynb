```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

### Human Coded Result Load

```python
with open('position_human_code.pkl', 'rb') as archivo:
    position_human = pickle.load(archivo)
with open('orientation_human_code.pkl', 'rb') as archivo:
    orientation_human = pickle.load(archivo)
with open('velocity_human_code.pkl', 'rb') as archivo:
    velocity_human = pickle.load(archivo)
with open('reward_human_code.pkl', 'rb') as archivo:
    reward_human = pickle.load(archivo)
```

### AI Coded Result Load

```python
with open('position_LLM_code.pkl', 'rb') as archivo:
    position_LLM = pickle.load(archivo)
with open('orientation_LLM_code.pkl', 'rb') as archivo:
    orientation_LLM = pickle.load(archivo)
with open('velocity_LLM_code.pkl', 'rb') as archivo:
    velocity_LLM = pickle.load(archivo)
with open('reward_LLM_code.pkl', 'rb') as archivo:
    reward_LLM = pickle.load(archivo)
```

### Visualization of results

```python
plt.plot(reward_human, label="Human coded", color="blue", linewidth=2, linestyle='-',)
plt.plot(reward_LLM, label="LLM coded", color="red", linewidth=2, linestyle='--',)
plt.grid()
plt.title("Evolution of cumulative reward by model")
plt.xlabel("Frames")
plt.ylabel("Reward")
plt.legend()
plt.savefig('reward_human_LLM.png')
files.download('reward_human_LLM.png')
```

```python
plt.plot(position_human, label="Human coded", color="blue", linewidth=2, linestyle='-',)
plt.plot(position_LLM, label="LLM coded", color="red", linewidth=2, linestyle='--',)
plt.grid()
plt.title("Evolution of X-position by model")
plt.xlabel("Frames")
plt.ylabel("X-position (m)")
plt.legend()
plt.savefig('position_human_LLM.png')
files.download('position_human_LLM.png')
```

```python
plt.plot(velocity_human, label="Human coded", color="blue", linewidth=2, linestyle='-',)
plt.plot(velocity_LLM, label="LLM coded", color="red", linewidth=2, linestyle='--',)
plt.grid()
plt.title("Evolution of X-velocity by model")
plt.xlabel("Frames")
plt.ylabel("X-velocity (m/s)")

plt.legend()
plt.savefig('velocity_human_LLM.png')
files.download('velocity_human_LLM.png')
```

```python
orientation_human=[i*180/np.pi for i in orientation_human]
orientation_LLM=[i*180/np.pi for i in orientation_LLM]
```

```python
plt.plot(orientation_human, label="Human coded", color="blue", linewidth=2, linestyle='-',)
plt.plot(orientation_LLM, label="LLM coded", color="red", linewidth=2, linestyle='--',)
plt.grid()
plt.title("Evolution of X-orientation by model")
plt.xlabel("Frames")
plt.ylabel("X-orientation (º)")

plt.legend()
plt.savefig('orientation_human_LLM.png')
files.download('orientation_human_LLM.png')
```
